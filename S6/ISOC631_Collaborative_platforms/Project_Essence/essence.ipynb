```python
from datetime import datetime
import requests
import zipfile
import os
import xml.etree.ElementTree as ET
```

```python
def get_zip_file():

    if(os.path.exists("old_price.xml")):
        os.remove("old_price.xml")

    if(os.path.exists("new_price.xml")):
        os.rename("new_price.xml", "old_price.xml")

    zip_file = datetime.now().strftime("%d-%m-%Y %Hh") + ".zip"

    url = requests.get("https://donnees.roulez-eco.fr/opendata/instantane")
    open(zip_file, "wb").write(url.content)

    with zipfile.ZipFile(zip_file, 'r') as in_zip:
        in_zip.extractall("")
        os.rename("PrixCarburants_instantane.xml","new_price.xml" )
        

get_zip_file()
```

```python
tree = ET.parse('new_price.xml')
root = tree.getroot()

for child in root:
    if child.find('prix') is not None and child.find('prix').attrib['id'] == '1':
        nom = child.find('prix').attrib['nom']
        valeur = child.find('prix').attrib['valeur']

    adresse = child.find('adresse').text   
    ville = child.find('ville').text  
    cp = child.attrib['cp']
    latitude = child.attrib['latitude']
    longitude = child.attrib['longitude']

    print(nom + "\n" + valeur + "\n" + adresse + "\n" + ville + "\n" + cp + "\n" + latitude + "\n" + longitude + "\n")
```
